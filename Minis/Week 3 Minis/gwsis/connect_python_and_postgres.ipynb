```python
import psycopg2
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
user = 'postgres'
password = 'postgres'
host = 'localhost'
port = 5432  
dbname = 'gwsis'

engine = create_engine(f'postgres://{user}:{password}@{host}:{port}/{dbname}')
conn = engine.connect()
```

```python
student_data = pd.read_excel('Resources/student_list.xlsx')
student_data.head()
```

```python
staff_data = pd.read_excel('Resources/staff_list.xlsx')
staff_data.head()
```

```python
course_data = pd.read_excel('Resources/course_list.xlsx')
course_data.head()
```

```python
student_data.to_sql('student', conn, if_exists='append', index=False)
# Options include: replace, fail, append
# index=False avoids adding DataFrame index to SQL database
```
